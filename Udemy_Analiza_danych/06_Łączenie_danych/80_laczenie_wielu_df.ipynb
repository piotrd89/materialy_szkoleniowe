# Zadanie

```python
# Corrected 'SupllierID' to 'SupplierID'
suppliers = pd.read_csv('/data/workspace_files/suppliers.csv', usecols=["SupplierID", "CompanyName", "City"])
suppliers.head()
```

```python
products = pd.read_csv('/data/workspace_files/products.csv',
                       usecols=['ProductID','ProductName','SupplierID','CategoryID'])
products.head()
```

```python
categories = pd.read_csv('/data/workspace_files/categories1.csv',
                       usecols=['CategoryID','CategoryName'], index_col='CategoryID')
categories.head()
```

```python
prod_cat = products.merge(categories, left_on='CategoryID', right_index=True)
prod_cat.head()
```

```python
prod_cat_sup = prod_cat.merge(suppliers, left_on='SupplierID', right_on='SupplierID')
prod_cat_sup.head()
```

```python
products.merge(categories, left_on="CategoryID", right_index=True
               ).merge(suppliers, left_on="SupplierID", right_on="SupplierID").head()
```
