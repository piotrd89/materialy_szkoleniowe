# Wstęp

```python
import pandas as pd
import numpy as np
```

```python
cat1 = pd.read_csv('/data/workspace_files/categories_del_1.csv',
                   usecols=['CategoryID','CategoryName'])
cat1
```

```python
cat2 = pd.read_csv('/data/workspace_files/categories_del_2.csv',
                   usecols=['CategoryID','CategoryName'])
cat2
```

```python
cat1.merge(cat2)
```

```python
cat1.merge(cat2,on='CategoryID')
```

```python
cat1.merge(cat2, on='CategoryID', suffixes=['_1','_2'])
# Paremetr suffixes pozwala nadawać numery do nazw kolumn, które znajdują się zarówno po lewej jak i praej stronie złączenia
```

```python
cat1.merge(cat2, on='CategoryID', suffixes=['_1','_2'],how='outer')
```

```python
cat1.merge(cat2, on='CategoryID', suffixes=['_1','_2'],how='outer', indicator=True)
# parametr indicator dodaje informację skąd wziął się dany wiersz
```

```python
cat_merged = cat1.merge(cat2, on='CategoryID', suffixes=['_1','_2'], how='outer', indicator=True)
```

```python
filter = cat_merged['_merge'] == 'left_only'
cat_merged[filter]
```

```python
cat1.merge(cat2, on='CategoryID', suffixes=['_1','_2'], how='left', indicator=True)
```

```python
cat1.merge(cat2, on='CategoryID', suffixes=['_1','_2'], how='right', indicator=True)
```

# Zadanie

```python
f_2015 = pd.read_csv('/data/workspace_files/Fortune_500_2015.csv', 
                     usecols=['rank','company'])
f_2015.head()
```

```python
f_2016 = pd.read_csv('/data/workspace_files/Fortune_500_2016.csv', 
                     usecols=['rank','company'])
f_2016.head()
```

```python
companies_2015_2016 = f_2015.merge(f_2016, on='company', how='inner', suffixes=['_2015','_2016'])
companies_2015_2016
```

```python
companies_2015_2016[companies_2015_2016['company'] == 'Microsoft']
```

```python
companies_2015_2016 = f_2015.merge(f_2016, on='company', how='outer', suffixes=['_2015','_2016'], indicator=True)
companies_2015_2016
```

```python
only_2015 = companies_2015_2016[companies_2015_2016['_merge'] == 'left_only']
only_2015
```

```python
only_2016 = companies_2015_2016[companies_2015_2016['_merge'] == 'right_only']
only_2016
```
