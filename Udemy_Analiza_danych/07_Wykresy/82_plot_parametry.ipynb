```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline

'''
%matplotlib inline  - pozwala na rysowanie wykresów w notatniku
'''
```

```python
bar = pd.read_csv('http://piotrdomagala.pl/data/weather_barcelona.csv', index_col='Date')
```

```python
bar.head()
```

```python
rom = pd.read_csv('http://piotrdomagala.pl/data/weather_rome.csv', index_col='Date')
rom.head()
```

```python
ams = pd.read_csv('http://piotrdomagala.pl/data/weather_amsterdam.csv', index_col='Date')
ams.head()
```

```python
temp_max = pd.DataFrame(index=bar.index)
temp_max['Barcelona'] = bar['TempMax']
temp_max['Rome'] = rom['TempMax']
temp_max['Amsterdam'] = ams['TempMax']
```

```python
temp_max.head()
```

```python
temp_max.plot()
```

```python
temp_max.plot(figsize=(15,5))
```

```python
temp_max.plot(figsize=(15,5), subplots=True)
```

```python
temp_max.plot(figsize=(15,5), use_index=False)

# wykorzystanie parametru use_index pozwala na to, by zmienić etykiety na osi X do kolejnych wierszy
```

```python
temp_max.plot(figsize=(15,5), title='Max temp w miastach')
```

```python
temp_max.plot(figsize=(15,5), legend=False)
```

```python
temp_max.plot(figsize=(15,5), logy=True)

# parametr logy - zmieniamy skalę wykresu, pozwala 'spłaszczyć' wyniki
# pozwala na porównanie wartości małych z dużymi
```

```python
temp_max[:14].plot(figsize=(15,5), xticks=(range(14)), rot=45)

# xticks - pozwala dla wartości z podanego zakresu, wykreślić etykiety na osi X
# rot - określa kąt pod jakim będą wyśwetlone etykiety na osi X
```

```python
temp_max[:7].plot(figsize=(15,5), table=True)

# table - wyświetla tabelę z danymi przedstawionymi na wykresie
```

```python
import datetime
#Import data
nasa = pd.read_csv("/data/workspace_files/nasa_facebook_statuses.csv", 
 usecols=["status_published","num_reactions", 
 "num_comments","num_shares","num_likes"])
#Convert column type to date time
nasa["status_published"] = pd.to_datetime(nasa["status_published"])
#Select observations between two datetimes - May 2015
filter =(nasa['status_published'] >= '2016-5-1') & (nasa['status_published'] < '2016-6-1')
nasa = nasa[filter]
#Add column with day only
nasa['day'] = nasa.apply(lambda row: row["status_published"].day,axis=1)
#Remove non-numeric column
nasa.drop('status_published', axis=1, inplace=True)
#Group all columns by sum
nasa_by_day = nasa.groupby(by='day').sum()
nasa_by_day.head()
```

```python
nasa_by_day.plot(figsize=(15,5))
```

```python
nasa_by_day.plot(figsize=(15,5), title='NASA Facebook activity')
```

```python
nasa_by_day.plot(figsize=(15,5), 
                 title='NASA Facebook activity',
                 xlabel = 'day',
                 subplots=True)
```

```python
nasa_by_day.plot(figsize=(15,5), 
                 title='NASA Facebook activity',
                 xlabel = 'day',
                 logy=True)
```

```python
nasa_by_day[:30].plot(figsize=(15,5), 
                 title='NASA Facebook activity',
                 xlabel = 'day',
                 logy=True,
                 rot=45
                      )
```

```python
nasa_by_day[:30].plot(figsize=(15,5), 
                 title='NASA Facebook activity',
                 xlabel = 'day',
                 logy=True,
                 rot=45,
                      table=True
                      )
```
