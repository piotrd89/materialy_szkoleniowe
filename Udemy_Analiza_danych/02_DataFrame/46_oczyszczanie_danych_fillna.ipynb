# Wstęp

```python
import pandas as pd, numpy as np
frame = pd.read_csv('http://piotrdomagala.pl/data/Mcdonalds.csv')
```

```python
frame.loc[1,'TotalFat'] = np.NaN
frame.loc[1,'Saturated Fat'] = np.NaN
frame.loc[2,'Saturated Fat'] = np.NaN
frame.loc[3,'Category'] = np.NaN
```

```python
frame.head()
```

```python
frame.fillna(value=0).head()
```

```python
# możemy przypisać wartości do podmiany dla każdej kolumny osobno

replaceDefinitionTable = {'Category':'UNKNOWN', 'TotalFat':0,'SaturatedFat':0}
frame.fillna(value=replaceDefinitionTable).head()
```

```python
# wypełniamy puste wartości dla konkretnej kolumny

frame['Category'].fillna(value='UNKNOWN', inplace=True)
frame.head()
```

```python
frame['Saturated Fat'].fillna(value=0, inplace=True)
frame.head()
```

```python
frame['TotalFat'].fillna(value=0, inplace=True)
frame.head()
```

```python
frame.loc[1,'TotalFat'] = np.NaN
frame.loc[1,'Saturated Fat'] = np.NaN
frame.loc[2,'Saturated Fat'] = np.NaN
frame.loc[3,'Category'] = np.NaN
```

```python
frame.head()
```

```python
avgTotalFat = frame['TotalFat'].mean()
avgTotalFat
```

```python
frame['TotalFat'].fillna(avgTotalFat, inplace=True)
frame.head()
```

```python
frame['Saturated Fat'].fillna(method='ffill', inplace=True)
frame.head()
```

```python
import pandas as pd, numpy as np
fuel = pd.read_csv('http://piotrdomagala.pl/data/fuel.csv',
                   low_memory=False,
                   usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type', 'Combined MPG (FT1)'],
                   index_col='Vehicle ID'
                   )
```

```python
fuel.loc[27705, 'Class'] = np.NaN
fuel.loc[26561, 'Class'] = np.NaN
fuel.loc[27550, 'Fuel Type'] = np.NaN
fuel.loc[27705, 'Combined MPG (FT1)'] = np.NaN
fuel.loc[27681, 'Combined MPG (FT1)'] = np.NaN

fuel.head()
```

```python
fuel.fillna(value=-1).head()
```

```python
replaceRules = {'Class':'---', 'Fuel Type':'---', 'Combined MPG (FT1)':-1}
fuel.fillna(value=replaceRules).head()
```

```python
avgMPG = fuel['Combined MPG (FT1)'].mean()
avgMPG
```

```python
fuel['Class'].fillna(value='?', inplace=True)
fuel['Fuel Type'].fillna(value='?', inplace=True)
fuel['Combined MPG (FT1)'].fillna(value=-1, inplace=True)

fuel.head()
```

```python
fuel.loc[27705, 'Class'] = np.NaN
fuel.loc[26561, 'Class'] = np.NaN
fuel.loc[27550, 'Fuel Type'] = np.NaN
fuel.loc[27705, 'Combined MPG (FT1)'] = np.NaN
fuel.loc[27681, 'Combined MPG (FT1)'] = np.NaN

fuel.head()
```

```python
fuel['Combined MPG (FT1)'].fillna(method='ffill', inplace=True)
fuel.head()
```
