# Wstęp

```python
import pandas as pd
import numpy as np

frame = pd.read_csv('http://piotrdomagala.pl/data/Mcdonalds.csv', 
                    usecols=['Item', 'Category', 'Serving Size', 'Calories', 'TotalFat'])
```

```python
frame['Calories'].rank().head()
```

```python
frame['CaloriesRank'] = frame['Calories'].rank()
frame.head()
```

```python
frame.sort_values(by='Calories').head(20)
```

```python
frame['CaloriesRank'] = frame['Calories'].rank(ascending=False)
frame.head()
```

```python
frame.sort_values(by='Calories', ascending=False).head(20)
```

```python
frame.sort_values(by='Calories', ascending=False, inplace=True)
frame.head(3)
```

```python
frame.nlargest(n=3, columns='Calories')
```

```python
frame.tail(3)
```

```python
frame.nsmallest(n=3, columns='Calories')
```
