# Wstęp

```python
import pandas as pd, numpy as np
frame = pd.read_csv('http://piotrdomagala.pl/data/Mcdonalds.csv')
```

```python
frame.loc[1,'Calories']= np.NaN
frame.loc[2,'TotalFat'] = np.NaN

frame.head()
```

```python
frame.dropna().head()
```

```python
# how=any - usuwa z ramki wszystkie wiersze, gdzie chociaż jedno pole jest puste (wartośc domyślna)
# how=all - usuwa z ramki wszystkie wiersze, gdzie wszystkie pola są puste

frame.dropna(how="all").head()
```

```python
# subset określa zakres w jakim wartości nie mogą być puste i usuwa wiersze z wartościami pustymi we wskazanych kolumnach
frame.dropna(subset=['TotalFat'], how='any').head()
```

```python
# axis określa czy bierzemy pod uwagę wierszy czy kolumny
# axis='columns' (1) usuwa kolumny w których znalezione zostaly puste wartości
# axis='rows' (0) usuwa wiersze, w których znalezione zostaly puste wartości
frame.dropna(axis='rows').head()
```

```python
# inplace - usuwamy wartości na oryginalnej ramce danych
frame.dropna(inplace=True)
frame
```
