# Wstęp

```python
import pandas as pd
```

```python
frame = pd.read_csv('http://piotrdomagala.pl/data/Mcdonalds.csv', usecols=['Item', 'Category', 'Calories', 'TotalFat', 'Sugars', 'Protein', 'Cholesterol'])
```

```python
frame.head()
```

```python
frame.TotalFat.head()
```

```python
type(frame.TotalFat.head())
```

```python
frame['TotalFat'].head()
```

```python
frame['TotalFat0'] = 0
```

```python
frame.head()
```

```python
SugarAndFat = frame.Sugars + frame.TotalFat
frame["SugarAndFat"] = SugarAndFat
frame.head()
```

```python
frame['SugarAndProtein'] = frame.Sugars + frame.Protein
frame.head()
```

```python
# wstawiamy kolumnę w określonym miejscu w DataFrame
frame.insert(loc=2, column='SPF', value = frame.Sugars + frame.Protein + frame.TotalFat)
frame.head()
```

```python
# usuwamy określone kolumny
frame = frame.drop(columns=['TotalFat0'])
```

```python
frame.head()
```

```python
frame.drop(labels="SugarAndFat", axis=1, inplace=True)
```

```python
frame.head()
```

```python
frame.drop(columns=['SPF'], inplace=True)
frame.head()
```

```python
import pandas as pd
```

```python
fuel = pd.read_csv('http://piotrdomagala.pl/data/fuel.csv', usecols=['Vehicle ID', 'Year', 'Make', 'Model', 'Class', 'Fuel Type', 'Combined MPG (FT1)'], low_memory=False)
fuel.head()
```

```python
fuel['Combined KPL'] = fuel['Combined MPG (FT1)'] * 1.609/3.78
fuel.head()
```

```python
fuel.insert(loc=4, column='Liters per km', value=100/fuel['Combined KPL'])
```

```python
fuel.head()
```

```python
del fuel["Combined MPG (FT1)"]
fuel.head()
```

```python
fuel.drop(columns='Combined KPL', inplace = True)
fuel.head()
```
